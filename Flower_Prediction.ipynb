```
import pandas as pd
```

```
import numpy as np
```

```
df=pd.read_csv("/content/drive/MyDrive/iris.data",names=["sepal length in cm",
"sepal width in cm",
"petal length in cm",
"petal width in cm","class"
],)
```

```
df.head()
```

```
df.info()
```

```
xtrain=df.drop("class",axis=1)
```

```
xtrain.iloc[15]
```

```
ytrain=df["class"]
```

```
ytrain
```

```
from sklearn.model_selection import train_test_split
```

```
x_train,x_test,y_train,y_test=train_test_split(xtrain,ytrain,test_size=0.25,random_state=5)
```

```
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```
from sklearn.tree import DecisionTreeClassifier
```

```
model=DecisionTreeClassifier()
```

```
model.fit(x_train,y_train)
```

```
model.score(x_train,y_train)
```

```
model.score(x_test,y_test)
```

```
from sklearn.model_selection import GridSearchCV
```

```
param={"criterion":['gini',"entropy"],"max_depth":[3,5,10,20,30,40,50],
       "min_samples_leaf":[3,5,10,20,30,40,50,100],
       "min_samples_split":[10,20,30,40,50,70],
       }
```

```
grid=GridSearchCV(model,param,cv=5,verbose=2)
```

```
grid.fit(x_train,y_train)
```

```
grid.best_estimator_
```

```
model1=DecisionTreeClassifier(ccp_alpha=0.0, class_weight=None, criterion='gini',
                       max_depth=3, max_features=None, max_leaf_nodes=None,
                       min_impurity_decrease=0.0, min_impurity_split=None,
                       min_samples_leaf=3, min_samples_split=40,
                       min_weight_fraction_leaf=0.0, presort='deprecated',
                       random_state=None, splitter='best')

```

```
model1.fit(x_train,y_train)
```

```
model1.score(x_train,y_train)
```

```
model1.score(x_test,y_test)
```

```
model1.predict([[5.7,4.4,1.5,0.3]])
```

```
ytrain[15]
```

```
yp=model1.predict(x_test)
```

```
from sklearn.metrics import confusion_matrix,f1_score,recall_score,precision_score
```

```
confusion_matrix(y_test,yp)
```

```
recall_score(y_test,yp,average=None)
```

```
f1_score(y_test,yp,average=None)
```
